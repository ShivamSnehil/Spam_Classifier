```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SMSSpamCollection.txt', sep='\t',names=['Category','Message'])
df.head(15)
```

```python
data  = df.where((pd.notnull(df)),'')
data.head(15)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['Category'] = encoder.fit_transform(data['Category'])
data.head(15)
```

```python
X = data['Message']
Y = data['Category']
```

```python
!pip install nltk
import re
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lematizer = WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(X)):
    review=re.sub('[^a-zA-z]',' ',X[i])
    review=review.lower()
    review=review.split()
    review=[lematizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus[:14]
```

```python
!pip install gensim
import gensim
from gensim.models import Word2Vec,KeyedVectors
from gensim.utils import simple_preprocess
```

```python
data['Message_list'] = corpus
data.head(15)
```

```python
data['Message_list1'] = data['Message_list'].apply(lambda x: gensim.utils.simple_preprocess(x))
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data['Message_list1'],data['Category'], test_size=0.2)
X_train = X_train.reset_index(drop = True)
X_test = X_test.reset_index(drop = True)
```

```python
w2v_model = gensim.models.Word2Vec(X_train,vector_size=100,window=5,min_count=1)
```

```python
words = set(w2v_model.wv.index_to_key)

# Compute average Word2Vec vectors
def avg_word2vec_transform(tokenized_message, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    n_words = 0
    for word in tokenized_message:
        if word in model.wv:
            feature_vector = np.add(feature_vector, model.wv[word])
            n_words += 1
    if n_words > 0:
        feature_vector = np.divide(feature_vector, n_words)
    return feature_vector

X_train_avg = [avg_word2vec_transform(ls, w2v_model, w2v_model.vector_size) for ls in X_train]
X_test_avg = [avg_word2vec_transform(ls, w2v_model, w2v_model.vector_size) for ls in X_test]
```

```python
X_test_avg[0]
```

```python
Y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc_classifier = rfc.fit(X_train_avg, Y_train)
```

```python
Y_pred = rfc_classifier.predict(X_test_avg)
```

```python
from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
accuracy = (Y_pred==Y_test).sum()/len(Y_pred)

precision = precision_score(Y_test, Y_pred, pos_label=0)
recall = recall_score(Y_test, Y_pred, pos_label = 0)

print('Accuracy: {} , Precision: {} , Recall: {}'.format(accuracy, precision, recall))
print(classification_report(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))
```

```python
# Function to preprocess new input email
def preprocess_new_email(email):
    review = re.sub('[^a-zA-Z]', ' ', email)
    review = review.lower()
    review = review.split()
    review = [lematizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    return review

# Function to predict whether a new email is spam or ham
def predict_email(email, model, w2v_model, vector_size):
    processed_email = preprocess_new_email(email)
    email_avg_w2v = avg_word2vec_transform(processed_email, w2v_model, vector_size).reshape(1, -1)
    prediction = model.predict(email_avg_w2v)
    return 'Ham mail' if prediction[0] == 0 else 'Spam mail'
```

```python
# Take input email from the user
input_email = input("Enter your email: ")
result = predict_email(input_email, rfc_classifier, w2v_model, w2v_model.vector_size)
print("Prediction:", result)
```
